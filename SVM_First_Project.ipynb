## IMPORT LIBRARIES AND DATA

```python
# Import the pandas library for data manipulation
import pandas as pd

# Import the function to load the Iris dataset from scikit-learn
from sklearn.datasets import load_iris

# Load the Iris dataset into a variable
iris = load_iris()

# Create a DataFrame from the iris data with column names from feature_names
df = pd.DataFrame(iris.data, columns=iris.feature_names)
# Displays the first 5 rows of the DataFrame
df.head()
```

## TARGETS

```python
# Add a new comumn 'target' to the DataFrame containing the class labels
df['target'] = iris.target
# Display the first 5 rows to the updated DataFrame
df.head()
```

```python
# Displays the target class names (species of iris) corresponding to the numeric labels in 'target'
iris.target_names
```

```python
# Displays the first 5 rows of the DataFrame where the target label is 2 (corresponding to the species 'virginica')
df[df.target==2].head()
```

```python
# Add a new column with the flower name for each numeric target label
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])

# Display the first five rows of the updated DataFrame
df.head()
```

```python
# Import the function to split the dataset into a training and testing sets
from sklearn.model_selection import train_test_split
```

```python
# Create the feature matrix X by dorpping the target and flower name colmuns (not needed for training)
X = df.drop(['target','flower_name'], axis = 'columns')
```

```python
# Define the target variable y as the 'target' column from the DataFrame
y = df.target
```

```python
# Split the data into training and testing sets: 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Import the support vector classifier model from scikit-learn
from sklearn.svm import SVC

# Create an instance of the SVC model
model = SVC()
```

```python
# Train the scv model sing the training data
model.fit(X_train, y_train)
```

```python
# Evaluate the models accuracy on the testing data
model.score(X_test, y_test)
```
